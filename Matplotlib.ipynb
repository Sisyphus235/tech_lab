````python
# import norms and abbreviation
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
````

````python
t = np.arange(0, 2.0, 0.01) # arange([start,]stop,[step,]dtype=None), returns ndarray
# print(t)
s = 1 + np.sin(2 * np.pi * t) # sin(x, /, out=None, *, where=True, casting='same_kind', order='K', dtype=None, subok=True[, signature, extobj])
# print(s)
fig, ax = plt.subplots() # subplots(nrows=1, ncols=1, sharex=False, sharey=False, squeeze=True, subplot_kw=None, gridspec_kw=None, **fig_kw)
# returns 1.fig: matplotlib.figure.Figure object, 2.ax: Axes object or array of Axes objects
# print(fig, ax, sep='::')
ax.plot(t, s)
# print(dir(fig), '\n', dir(ax))b

ax.set(xlabel='time(s)', ylabel='voltage(mV)', title='Simple test')
ax.grid()

plt.show()
````
